```python
import re
lines = []
with open('/content/drive/MyDrive/Website/websites.txt') as f:
  lines = f.readlines()

r = re.compile("^((?:http(?:s)?\:\/\/)?[a-zA-Z0-9_-]+(?:.[a-zA-Z0-9_-]+)*.[a-zA-Z]{2,4}(?:\/[a-zA-Z0-9_]+)*(?:\/[a-zA-Z0-9_]+.[a-zA-Z]{2,4}(?:\?[a-zA-Z0-9_]+\=[a-zA-Z0-9_]+)?)?(?:\&[a-zA-Z0-9_]+\=[a-zA-Z0-9_]+)*)$")
newlist = list(filter(r.match, lines))

final_url = []
for url in newlist:
  url1 = url.split()
  final_url.append("https://"+ url1[0])

print(final_url)
print("The total number of items: ",len(lines))
print("The total number of websites found", len(final_url))
```

```python
import requests
from bs4 import BeautifulSoup

for url2 in final_url:
  try:
    reqs = requests.get(url2)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    print("Title of the website is :")
    for title in soup.find_all('title'):
      print(url2, title.get_text())
  except:
    print("Website not found for: \n", url2)
```
